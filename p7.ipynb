# Text Analytics
### 1. Extract Sample document and apply following document preprocessing methods: Tokenization, POS Tagging, stop words removal, Stemming and Lemmatization.
### 2. Create representation of document by calculating Term Frequency and Inverse Document Frequency.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_fake = pd.read_csv('Fake.csv')
df_fake['true'] = 0
df_fake.head()
```

```python
df_true = pd.read_csv('True.csv')
df_true['true'] = 1
df_true.head()
```

```python
df = pd.concat([df_true, df_fake], ignore_index=True)
df = df.sample(frac=1).reset_index(drop=True)
df.head()
```

```python
df.drop(columns = ['subject', 'date'], inplace = True)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df[df.duplicated() == True]
# df[df['title'] == df.iloc[900]['title']]
```

```python
df = df.drop_duplicates()
```

```python
df.shape
```

```python
df['title'] = df.title.str.lower()
df['text'] = df.text.str.lower()
```

```python
df.head()
```

```python
df = df.drop(columns=['text'])
```

```python
import nltk

nltk.download('punkt')

df['title_sent_token'] = df['title'].apply(nltk.sent_tokenize)
df['title_word_token'] = df['title'].apply(nltk.word_tokenize)

df.head()
```

```python
nltk.download('averaged_perceptron_tagger')

df['title_word_pos'] = df['title_word_token'].apply(nltk.pos_tag)
df.head()
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

df['title_word_token_stopwords'] = df['title_word_token'].apply(lambda x: [word for word in x if word not in stop_words])

df.head()
```

```python
nltk.download('wordnet')

df['title_word_stem'] = df['title_word_token_stopwords'].apply(lambda x: [nltk.PorterStemmer().stem(word) for word in x])
df['title_word_lem'] = df['title_word_token_stopwords'].apply(lambda x: [nltk.WordNetLemmatizer().lemmatize(word) for word in x])

df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()

tfidf_matrix = vectorizer.fit_transform([' '.join(words) for words in df['title_word_lem']])

tfidf_matrix.shape
```
