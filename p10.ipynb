```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
iris = fetch_ucirepo(id=53)

# data (as pandas dataframes)
X = iris.data.features
y = iris.data.targets
```

```python
X.head()
```

```python
df = pd.merge(X, y, left_index=True, right_index=True)
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

### List down the features and their types (e.g., numeric, nominal) available in the dataset.

```python
df.info()
```

- `petal length` : Numeric Continuous
- `petal width` : Numeric Continuous


### Create a histogram for each feature in the dataset to illustrate the feature distributions.

```python
for col in df.columns:
    sns.histplot(df[col], bins = 20)
    plt.xlabel(col)
    plt.show()
```

### Create a boxplot for each feature in the dataset.

```python
sns.boxplot(df)
```
